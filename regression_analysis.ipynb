```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
from functools import reduce
from tqdm import tqdm
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn import linear_model
from matplotlib import pyplot as plt
```

```python
def create_dummy_vars(df, columns):
    for column in columns:
        dummy_vars = columns[column]
        for dummy_var in dummy_vars:
            var_value = dummy_vars[dummy_var]
            df.loc[(df[column].isin(var_value)), dummy_var] = 0
            df.loc[~(df[column].isin(var_value)), dummy_var] = 1
    return df

def calculate_vif(independet_df):
    vif_list = []
    x_cols = independet_df.columns
    for x_col in tqdm(x_cols):
        x = independet_df[[x_col]]
        y = independet_df.drop(columns = x_col)
        rsq = smf.ols(formula = 'x~y', data = independet_df).fit().rsquared
        vif = round(1/(1-rsq), 2)
        if vif > 2.5:
            threshold_flag = '1'
        elif vif <= 2.5:
            threshold_flag = '0'
        var_dict = {'independent_var': x_col, 'vif': vif, 'threshold': threshold_flag}
        vif_list.append(var_dict)
    vif_df = pd.DataFrame(vif_list)
    return vif_df

def create_formula(independent_vars):
    dummy_vars = str(str(independent_vars)[2:-2].replace('\', \'', ' + '))
    regression_formula = ("denied ~ " + dummy_vars)
    return regression_formula

def convert_results_to_df(model):
    coef_dict = {'results': dict(model.params), 'col_name': 'coefficient'}
    std_error_dict = {'results': dict(model.bse), 'col_name': 'standard_error'}
    tvalues_dict = {'results': dict(model.tvalues), 'col_name': 'z_value'}
    pvalues_dict = {'results': dict(model.pvalues), 'col_name': 'p_value'}
    oddsratio_dict = {'results': dict(np.exp(model.params)), 'col_name': 'odds_ratio'}
    results_dicts = [coef_dict, std_error_dict, tvalues_dict, pvalues_dict, oddsratio_dict]
    dfs = []
    for results_dict in results_dicts:
        col_name = results_dict['col_name']
        df = pd.DataFrame.from_dict(results_dict['results'], orient = 'index').reset_index().rename(columns = {'index': 'variable_name', 0: col_name})
        dfs.append(df)
    results_df = reduce(lambda left, right: pd.merge(left, right, on = ['variable_name']), dfs)
    results_df.insert(1, 'pseudo_rsquared', model.prsquared)
    return results_df

def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],"{:.0%}".format(y[i]))
```

```python
from fairlearn.metrics import (
    demographic_parity_difference,
    equalized_odds_difference,
    false_positive_rate_difference,
    false_negative_rate_difference
)

# Assuming your DataFrame is named 'df' and contains the required columns
df = data


# Define the outcome column and protected attributes
outcome_column = 'denied'
protected_attributes = ['native', 'asian', 'black', 'pac_islander', 'female']

# Convert the 'denied' column to integer form (0 and 1)
df['denied'] = df['denied'].astype(int)

# Convert the columns to integer binary form (0 and 1)
for attr in protected_attributes:
    df[attr] = df[attr].astype(int)

# Prepare the data
y_true = df[outcome_column]  # True labels
y_pred = logit_model.predict(sm.add_constant(df[variables]))

# Define the threshold for converting probabilities to binary predictions
threshold = 0.5

# Convert probabilities to binary predictions
y_pred_binary = (y_pred >= threshold).astype(int)
```

Fairness Evaluation

```python
# Calculate Fairlearn metrics using binary predictions
dpd = demographic_parity_difference(y_true, y_pred_binary, sensitive_features=df[protected_attributes])
eod = equalized_odds_difference(y_true, y_pred_binary, sensitive_features=df[protected_attributes])
fpr = false_positive_rate_difference(y_true, y_pred_binary, sensitive_features=df[protected_attributes])
fnr = false_negative_rate_difference(y_true, y_pred_binary, sensitive_features=df[protected_attributes])

# Print the fairness metrics
print(f"Demographic Parity Difference: {dpd}")
print(f"Equalized Odds Difference: {eod}")
print(f"False Positive Rate Difference: {fpr}")
print(f"False Negative Rate Difference: {fnr}")
```

#### Bias Mitigation through postprocessing

```python
from fairlearn.postprocessing import ThresholdOptimizer
from fairlearn.metrics import equalized_odds_difference
from sklearn.linear_model import LogisticRegression


# Initialize a base estimator (for example, logistic regression)
#base_estimator = LogisticRegression()

X = df[variables]
# Create and fit a Logistic Regression model using scikit-learn
model = LogisticRegression(solver='liblinear')
model.fit(X, y_true)

# Get predicted probabilities
y_pred1 = model.predict_proba(X)[:, 1]

# Now you can pass the model to the ThresholdOptimizer


# Initialize the ThresholdOptimizer with the base estimator
optimizer = ThresholdOptimizer(
    estimator=model,  # Base estimator (classifier)
    constraints="equalized_odds",
    grid_size=1000
)

```

```python
# Fit the optimizer with input features, true labels, and sensitive features
optimizer.fit(
    X=X,
    y=y_true,
    sensitive_features=df[protected_attributes]
)

# Obtain adjusted predictions using the optimizer
y_pred_adjusted = optimizer.predict(
    X=X,  # The input features (data)
    sensitive_features=df[protected_attributes]  # The sensitive features
)

# Evaluate the equalized odds difference after adjustments
eod_adjusted = equalized_odds_difference(
    y_true, y_pred_adjusted, sensitive_features=df[protected_attributes]
)

# Print the equalized odds difference after adjustment
print(f"Equalized Odds Difference after adjustment: {eod_adjusted}")
```
